## Slicing and Extracting data 

```python
# Isolating one column using [ ] 
pokemon['Name']
```

```python
# Isolating two or more columns using [[ ]]
pokemon[['Name','HP']]
```

```python
# Isolating one row using [ ] 
# .index returns the row labels of the DataFrame
pokemon[pokemon.index==1]
```

```python
# Isolating two or more rows using [ ] 
pokemon[pokemon.index.isin(range(2,10))]
```

```python
pokemon
```

```python
# Conditional slicing 
# For example we want to look at pokemons that have attack>50

pokemon[pokemon['Attack']>50]
```

```python
# Add or replace a column

pokemon = pokemon.assign(Total=pokemon["HP"] + pokemon["Attack"] + pokemon["Defense"] + pokemon["Sp. Atk"] + pokemon["Sp. Def"] + pokemon["Speed"])
pokemon.head()
```

```python
# Drop a column

pokemon = pokemon.drop(columns=["Total"])
pokemon.head()
```

```python
# Set and reset the index

pokemon = pokemon.set_index("Name")
pokemon = pokemon.reset_index()
pokemon
```

```python
# Sort values
pokemon.sort_values(by="Attack", ascending=False)
```

```python
# Using iloc (by index position)
print(pokemon.iloc[0])  # First row

# Using loc (by index label)
pokemon = pokemon.set_index("Name")  # Set Name as the index
print(pokemon.loc["Bulbasaur"])
```

```python
# Group by and aggregate
pokemon.groupby("Type 1").mean()
```

```python
# Plotting
pokemon["HP"].plot(kind="hist", bins=10, density=True)
```
